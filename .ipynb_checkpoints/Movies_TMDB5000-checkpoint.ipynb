## Data Preprocessing

```python
# Creating a master datasets out of the given 2 tables by merging on on the common column 'title'
movies = pd.merge(movies, credits, on = 'title')
movies.shape
```

```python
movies.head(1)
```

```python
# Making a list of imp features to be used for the modelling
# genres
# movie_id
# keywords
# overview
# cast
# crew

movies = movies[['movie_id','title','genres','keywords','overview','cast','crew']]
```

```python
movies.head()
```

```python
# Checking null rows
movies.isnull().sum()
```

```python
# Dropping null rows
movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
# Dropping duplicate rows
movies.duplicated().sum()
```

```python
# Checking on an element in the genres column
movies.iloc[0,2]
```

```python
# Lets extract the names from the above string
# Since the elements of the genres column are in a '' form, the operation of removing data from this colmn requires applying a transformation 
# on each element of the column. The Transformation done by creating an ast.literal_eval object out of each element and using this object for any operations thereafter
```

```python
import ast
# Creating a helper function to apply transformations on the columns
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
# Applying same helper function on keywords column
movies['keywords'] = movies['keywords'].apply(convert)
movies['keywords']
```

```python
def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
            counter += 1
        else: 
            break
    return L
```

```python
# Only keeping the first 3 names in the cast column as the st 3 names are the most influencing names for an efficient algorithm. The 1st 3 names are the names of lead characters
movies['cast'] = movies['cast'].apply(convert2)
movies['cast']
```

```python
movies.head()
```

```python
# Observing an element of the crew column
movies['crew'][0]
```

```python
# Function to extract only the director's name from the crew column. Only the director's name is the most influencing element.
def convert3(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert3)
movies['crew']
```

```python
movies['overview'][0]
```

```python
# Splitting each element of the overview column into coma seperated values so that they can be furthur connected with other columns as a list of word.
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies['overview']
```

```python
type(movies['cast'][0])
```

```python
# Removing the spaces in between genres names. Eg 'Science Fiction' becomes 'ScienceFiction'. This will make the algorithm efficiently use the genres and other columns when using cosine similarity furthur in the code 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
# Increasing the number of words for genres and crew section
```

```python
x = []
# Assuming 'keywords' column contains lists
for keywords_list in movies['crew']:
    x.extend(keywords_list)
x_unique = list(set(x))
len(x_unique)
```

```python
len(x_unique)
```

```python
# Comboning all the columns to create a single column called tags
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew'] 
```

```python
# Creating a dataframe with only the 3 below mentioned columns
new_df = movies[['movie_id','title','tags']]
```

```python
# Removing all the special characters like commas, and dots to make it easier for the user to search movies.
new_df['title'] =  new_df['title'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df['tags'][0]
```

```python
new_df.head()
```

```python
# Now that our dataframe is ready, we shall use vectorization to find similarities amongst the movies.
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words = 'english', max_features = 5000)
vectors = cv.fit_transform(new_df['tags'])
vectors = vectors.toarray()
vectors
```

```python
vectors.shape
```

```python
# To get the words
X = cv.get_feature_names_out()
```

```python
np.set_printoptions(threshold=np.inf)
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
# Example of running PorterStemmer
ps.stem('organization')
```

```python
def stemming_function(text):
    U = []
    for i in text.split():
        U.append(ps.stem(i))
    return ' '.join(U)
```

```python
new_df['tags'] = new_df['tags'].apply(stemming_function)
```

```python
# Extracting the 1st element of the tags column
new_df['tags'][0]
```

```python
cv = CountVectorizer(stop_words = 'english', max_features = 5000)
vectors = cv.fit_transform(new_df['tags'])
vectors = vectors.toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
def recommend_similar_movies(name):
    movie_index = new_df[new_df['title'] == name].index[0]
    x = pd.DataFrame(enumerate(similarity[movie_index]))
    sorted_x = x.sort_values(by=1, ascending = False)
    top_movies_indexes = sorted_x.iloc[1:6,0]
    top_movies_list = top_movies_indexes.tolist()
    return new_df.iloc [top_movies_list,1]
```

```python
recommend_similar_movies('Avatar')
```
